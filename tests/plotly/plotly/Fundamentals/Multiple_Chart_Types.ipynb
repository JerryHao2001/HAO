
<h4 id="Line-Chart-and-a-Bar-Chart">Line Chart and a Bar Chart<a class="anchor-link" href="#Line-Chart-and-a-Bar-Chart">¶</a></h4>


```python

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=[0, 1, 2, 3, 4, 5],
        y=[1.5, 1, 1.3, 0.7, 0.8, 0.9]
    ))

fig.add_trace(
    go.Bar(
        x=[0, 1, 2, 3, 4, 5],
        y=[1, 0.5, 0.7, -1.2, 0.3, 0.4]
    ))

fig.show()

```


<h4 id="A-Contour-and-Scatter-Plot-of-the-Method-of-Steepest-Descent">A Contour and Scatter Plot of the Method of Steepest Descent<a class="anchor-link" href="#A-Contour-and-Scatter-Plot-of-the-Method-of-Steepest-Descent">¶</a></h4>


```python

import plotly.graph_objects as go

# Load data
import json
import six.moves.urllib

response = six.moves.urllib.request.urlopen(
    "https://raw.githubusercontent.com/plotly/datasets/master/steepest.json")

data = json.load(response)

# Create figure
fig = go.Figure()

fig.add_trace(
    go.Contour(
        z=data["contour_z"][0],
        y=data["contour_y"][0],
        x=data["contour_x"][0],
        ncontours=30,
        showscale=False
    )
)

fig.add_trace(
    go.Scatter(
        x=data["trace_x"],
        y=data["trace_y"],
        mode="markers+lines",
        name="steepest",
        line=dict(
            color="black"
        )
    )
)

fig.show()

```
