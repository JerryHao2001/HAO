
<h4 id="Offline-Configuration-Options">Offline Configuration Options<a class="anchor-link" href="#Offline-Configuration-Options">¶</a></h4><p>Now you can pass a <code>config</code> dictionary with all configurations options such as <code>scrollZoom</code>, <code>editable</code>, and <code>displayModeBar</code>. For the complete list of config options check out: <a href="https://github.com/plotly/plotly.js/blob/master/src/plot_api/plot_config.js">https://github.com/plotly/plotly.js/blob/master/src/plot_api/plot_config.js</a></p>
<h5 id="Enable-Scroll-Zoom">Enable Scroll Zoom<a class="anchor-link" href="#Enable-Scroll-Zoom">¶</a></h5>


```python

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=[1, 2, 3],
        y=[1, 3, 1]))

fig.show(config={'scrollZoom': True})

```


<h5 id="Display-ModeBar">Display ModeBar<a class="anchor-link" href="#Display-ModeBar">¶</a></h5>


```python

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=[1, 2, 3],
        y=[1, 3, 1]))

fig.show(config={'displayModeBar': True})

```


<h5 id="Edit-Mode---change-the-title-and-axis-titles">Edit Mode - change the title and axis titles<a class="anchor-link" href="#Edit-Mode---change-the-title-and-axis-titles">¶</a></h5>


```python

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=[1, 2, 3],
        y=[1, 3, 1]))

fig.show(config={'editable': True})

```


<h5 id="Multiple-Config-Options-at-Once!">Multiple Config Options at Once!<a class="anchor-link" href="#Multiple-Config-Options-at-Once!">¶</a></h5>


```python

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=[1, 2, 3],
        y=[1, 3, 1]))

fig.show(config={
    'scrollZoom': True,
    'displayModeBar': True,
    'editable': True
})

```


<h5 id="Remove-Modebar-Buttons">Remove Modebar Buttons<a class="anchor-link" href="#Remove-Modebar-Buttons">¶</a></h5>


```python

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=[1, 2, 3],
        y=[1, 3, 1]))

fig.show(config={
    'modeBarButtonsToRemove': ['toggleSpikelines','hoverCompareCartesian']
})

```
