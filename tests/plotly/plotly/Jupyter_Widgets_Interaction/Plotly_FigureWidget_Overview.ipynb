
<h4 id="Create-a-Simple-FigureWidget">Create a Simple FigureWidget<a class="anchor-link" href="#Create-a-Simple-FigureWidget">¶</a></h4><p>Create an empty FigureWidget and then view it.</p>


```

import plotly.graph_objects as go

f = go.FigureWidget()
f

```


<p><img src="https://raw.githubusercontent.com/michaelbabyn/plot_data/master/empty_fw.png"/></p>



<p>Add traces or update the layout and then watch the output above update in real time.</p>


```

f.add_scatter(y=[2, 1, 4, 3]);

```

```

f.add_bar(y=[1, 4, 3, 2]);

```

```

f.layout.title = 'Hello FigureWidget'

```


<p><img src="https://raw.githubusercontent.com/michaelbabyn/plot_data/master/figurewidget-create.gif"/></p>



<h4 id="Update-the-Data-and-the-Layout">Update the Data and the Layout<a class="anchor-link" href="#Update-the-Data-and-the-Layout">¶</a></h4>


```

# update scatter data
scatter = f.data[0]
scatter.y = [3, 1, 4, 3]

```

```

# update bar data
bar = f.data[1]
bar.y = [5, 3, 2, 8]

```

```

f.layout.title.text = 'This is a new title'

```


<h4 id="Construct-a-FigureWidget-from-a-Figure-graph-object">Construct a FigureWidget from a Figure graph object<a class="anchor-link" href="#Construct-a-FigureWidget-from-a-Figure-graph-object">¶</a></h4>



<p>A standard <code>Figure</code> object can be passed to the <code>FigureWidget</code> constructor.</p>


```

import plotly.graph_objects as go

trace = go.Heatmap(z=[[1, 20, 30, 50, 1], [20, 1, 60, 80, 30], [30, 60, 1, -10, 20]],
                   x=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
                   y=['Morning', 'Afternoon', 'Evening'])
data=[trace]
layout = go.Layout(title='Activity Heatmap')

figure = go.Figure(data=data, layout=layout)

f2 = go.FigureWidget(figure)
f2

```
