
<h2 id="Ternary-contour-plots">Ternary contour plots<a class="anchor-link" href="#Ternary-contour-plots">¶</a></h2>



<p>A ternary contour plots represents isovalue lines of a quantity defined inside a <a href="https://en.wikipedia.org/wiki/Ternary_plot">ternary diagram</a>, i.e. as a function of three variables which sum is constant. Coordinates of the ternary plot often correspond to concentrations of three species, and the quantity represented as contours is some property (e.g., physical, chemical, thermodynamical) varying with the composition.</p>
<p>For ternary contour plots, use the figure factory <code>create_ternary_contour</code>. The figure factory interpolates between given data points in order to compute the contours.</p>
<p>Below we represent an example from metallurgy, where the mixing enthalpy is represented as a contour plot for aluminum-copper-yttrium (Al-Cu-Y) alloys.</p>
<h4 id="Simple-ternary-contour-plot-with-plotly">Simple ternary contour plot with plotly<a class="anchor-link" href="#Simple-ternary-contour-plot-with-plotly">¶</a></h4>


```python

import plotly.figure_factory as ff
import numpy as np
Al = np.array([0. , 0. , 0., 0., 1./3, 1./3, 1./3, 2./3, 2./3, 1.])
Cu = np.array([0., 1./3, 2./3, 1., 0., 1./3, 2./3, 0., 1./3, 0.])
Y = 1 - Al - Cu
# synthetic data for mixing enthalpy
# See https://pycalphad.org/docs/latest/examples/TernaryExamples.html
enthalpy = (Al - 0.01) * Cu * (Al - 0.52) * (Cu - 0.48) * (Y - 1)**2
fig = ff.create_ternary_contour(np.array([Al, Y, Cu]), enthalpy,
                                pole_labels=['Al', 'Y', 'Cu'],
                                interp_mode='cartesian')
fig.show() 

```


<h4 id="Customized-ternary-contour-plot">Customized ternary contour plot<a class="anchor-link" href="#Customized-ternary-contour-plot">¶</a></h4>


```python

import plotly.figure_factory as ff
import numpy as np
Al = np.array([0. , 0. , 0., 0., 1./3, 1./3, 1./3, 2./3, 2./3, 1.])
Cu = np.array([0., 1./3, 2./3, 1., 0., 1./3, 2./3, 0., 1./3, 0.])
Y = 1 - Al - Cu
# synthetic data for mixing enthalpy
# See https://pycalphad.org/docs/latest/examples/TernaryExamples.html
enthalpy = 2.e6 * (Al - 0.01) * Cu * (Al - 0.52) * (Cu - 0.48) * (Y - 1)**2 - 5000
fig = ff.create_ternary_contour(np.array([Al, Y, Cu]), enthalpy,
                                pole_labels=['Al', 'Y', 'Cu'],
                                interp_mode='cartesian',
                                ncontours=20,
                                colorscale='Viridis',
                                showscale=True,
                                title='Mixing enthalpy of ternary alloy')
fig.show() 

```


<h4 id="Ternary-contour-plot-with-lines-only">Ternary contour plot with lines only<a class="anchor-link" href="#Ternary-contour-plot-with-lines-only">¶</a></h4>


```python

import plotly.figure_factory as ff
import numpy as np
Al = np.array([0. , 0. , 0., 0., 1./3, 1./3, 1./3, 2./3, 2./3, 1.])
Cu = np.array([0., 1./3, 2./3, 1., 0., 1./3, 2./3, 0., 1./3, 0.])
Y = 1 - Al - Cu
# synthetic data for mixing enthalpy
# See https://pycalphad.org/docs/latest/examples/TernaryExamples.html
enthalpy = 2.e6 * (Al - 0.01) * Cu * (Al - 0.52) * (Cu - 0.48) * (Y - 1)**2 - 5000
fig = ff.create_ternary_contour(np.array([Al, Y, Cu]), enthalpy,
                                pole_labels=['Al', 'Y', 'Cu'],
                                interp_mode='cartesian',
                                ncontours=20,
                                coloring='lines')
fig.show() 

```


<h4 id="Ternary-contour-plot-with-data-points">Ternary contour plot with data points<a class="anchor-link" href="#Ternary-contour-plot-with-data-points">¶</a></h4><p>With <code>showmarkers=True</code>, data points used to compute the contours are also displayed. They are best visualized for contour lines (no solid coloring). At the moment data points lying on the edges of the diagram are not displayed, this will be improved in future versions.</p>


```python

import plotly.figure_factory as ff
import numpy as np
Al, Cu = np.mgrid[0:1:7j, 0:1:7j]
Al, Cu = Al.ravel(), Cu.ravel()
mask = Al + Cu <= 1
Al, Cu = Al[mask], Cu[mask]
Y = 1 - Al - Cu

enthalpy = (Al - 0.5) * (Cu - 0.5) * (Y - 1)**2
fig = ff.create_ternary_contour(np.array([Al, Y, Cu]), enthalpy,
                                pole_labels=['Al', 'Y', 'Cu'],
                                ncontours=20,
                                coloring='lines',
                                showmarkers=True)
fig.show() 

```


<h4 id=" Interpolation-mode"> Interpolation mode<a class="anchor-link" href="# Interpolation-mode">¶</a></h4><p>Two modes are available in order to interpolate between data points: interpolation in Cartesian space (<code>interp_mode='cartesian'</code>) or interpolation using the <a href="https://link.springer.com/article/10.1023/A:1023818214614">isometric log-ratio transformation</a> (see also <a href="https://www.researchgate.net/profile/Leon_Parent2/post/What_is_the_best_approach_for_diagnosing_nutrient_disorders_and_formulating_fertilizer_recommendations/attachment/59d62a69c49f478072e9cf3f/AS%3A272541220835360%401441990298625/download/Egozcue+et+al+2003.pdf">preprint</a>),  <code>interp_mode='ilr'</code>. The <code>ilr</code> transformation preserves metrics in the <a href="https://en.wikipedia.org/wiki/Simplex">simplex</a> but is not defined on its edges.</p>


```python

a, b = np.mgrid[0:1:20j, 0:1:20j]
mask = a + b <= 1
a, b = a[mask], b[mask]
coords = np.stack((a, b, 1 - a - b))
value = np.sin(3.2 * np.pi * (a + b)) + np.sin(3 * np.pi * (a - b))
fig = ff.create_ternary_contour(coords, value, ncontours=9)
fig.show()

```
