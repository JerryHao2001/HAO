
<h3 id="Basic-Contour-Plot">Basic Contour Plot<a class="anchor-link" href="#Basic-Contour-Plot">¶</a></h3><p>A 2D contour plot shows the <a href="https://en.wikipedia.org/wiki/Contour_line">contour lines</a> of a 2D numerical array <code>z</code>, i.e. interpolated lines of isovalues of <code>z</code>.</p>


```python

import plotly.graph_objects as go

fig = go.Figure(data =
    go.Contour(
        z=[[10, 10.625, 12.5, 15.625, 20],
           [5.625, 6.25, 8.125, 11.25, 15.625],
           [2.5, 3.125, 5., 8.125, 12.5],
           [0.625, 1.25, 3.125, 6.25, 10.625],
           [0, 0.625, 2.5, 5.625, 10]]
    ))
fig.show()

```


<h3 id="Setting-X-and-Y-Coordinates-in-a-Contour-Plot">Setting X and Y Coordinates in a Contour Plot<a class="anchor-link" href="#Setting-X-and-Y-Coordinates-in-a-Contour-Plot">¶</a></h3>


<h3 id="Connect-the-Gaps-Between-None-Values-in-the-Z-Matrix">Connect the Gaps Between None Values in the Z Matrix<a class="anchor-link" href="#Connect-the-Gaps-Between-None-Values-in-the-Z-Matrix">¶</a></h3>


```python

import plotly.graph_objs as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=2, subplot_titles=('connectgaps = False',
                                                        'connectgaps = True'))
z = [[None, None, None, 12, 13, 14, 15, 16],
     [None, 1, None, 11, None, None, None, 17],
     [None, 2, 6, 7, None, None, None, 18],
     [None, 3, None, 8, None, None, None, 19],
     [5, 4, 10, 9, None, None, None, 20],
     [None, None, None, 27, None, None, None, 21],
     [None, None, None, 26, 25, 24, 23, 22]]

fig.add_trace(go.Contour(z=z, showscale=False), 1, 1)
fig.add_trace(go.Contour(z=z, showscale=False, connectgaps=True), 1, 2)
fig.add_trace(go.Heatmap(z=z, showscale=False, zsmooth='best'), 2, 1)
fig.add_trace(go.Heatmap(z=z, showscale=False, connectgaps=True, zsmooth='best'), 2, 2)

fig['layout']['yaxis1'].update(title='Contour map')
fig['layout']['yaxis3'].update(title='Heatmap')

fig.show()

```


<h3 id="Smoothing-the-Contour-lines">Smoothing the Contour lines<a class="anchor-link" href="#Smoothing-the-Contour-lines">¶</a></h3>


```python

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

z =   [[2, 4, 7, 12, 13, 14, 15, 16],
       [3, 1, 6, 11, 12, 13, 16, 17],
       [4, 2, 7, 7, 11, 14, 17, 18],
       [5, 3, 8, 8, 13, 15, 18, 19],
       [7, 4, 10, 9, 16, 18, 20, 19],
       [9, 10, 5, 27, 23, 21, 21, 21],
       [11, 14, 17, 26, 25, 24, 23, 22]]

fig = make_subplots(rows=1, cols=2,
                    subplot_titles=('Without Smoothing', 'With Smoothing'))

fig.add_trace(go.Contour(z=z, line_smoothing=0), 1, 1)
fig.add_trace(go.Contour(z=z, line_smoothing=0.85), 1, 2)

fig.show()

```


<h3 id="Color-Bar-Size-for-Contour-Plots">Color Bar Size for Contour Plots<a class="anchor-link" href="#Color-Bar-Size-for-Contour-Plots">¶</a></h3><p>In the example below, both the thickness (given here in pixels) and the length (given here as a fraction of the plot height) are set.</p>


```python

import plotly.graph_objects as go

fig = go.Figure(data=
    go.Contour(
        z=[[10, 10.625, 12.5, 15.625, 20],
           [5.625, 6.25, 8.125, 11.25, 15.625],
           [2.5, 3.125, 5., 8.125, 12.5],
           [0.625, 1.25, 3.125, 6.25, 10.625],
           [0, 0.625, 2.5, 5.625, 10]],
        colorbar=dict(
            thickness=25,
            thicknessmode='pixels',
            len=0.6,
            lenmode='fraction',
            outlinewidth=0
        )
    ))

fig.show()

```
