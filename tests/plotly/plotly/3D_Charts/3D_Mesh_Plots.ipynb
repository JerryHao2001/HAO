
<h3 id="Simple-3D-Mesh-example">Simple 3D Mesh example<a class="anchor-link" href="#Simple-3D-Mesh-example">¶</a></h3><p><code>go.Mesh3d</code> draws a 3D set of triangles with vertices given by <code>x</code>, <code>y</code> and <code>z</code>. If only coordinates are given, an algorithm such as <a href="https://en.wikipedia.org/wiki/Delaunay_triangulation">Delaunay triangulation</a> is used to draw the triangles. Otherwise the triangles can be given using the <code>i</code>, <code>j</code> and <code>k</code> parameters (see examples below).</p>


```

import plotly.graph_objects as go
import numpy as np

# Download data set from plotly repo
pts = np.loadtxt(np.DataSource().open('https://raw.githubusercontent.com/plotly/datasets/master/mesh_dataset.txt'))
x, y, z = pts.T

fig = go.Figure(data=[go.Mesh3d(x=x, y=y, z=z, color='lightpink', opacity=0.50)])
fig.show()

```


<h3 id="3D-Mesh-example-with-Alphahull">3D Mesh example with Alphahull<a class="anchor-link" href="#3D-Mesh-example-with-Alphahull">¶</a></h3>



<p>The <code>alphahull</code> parameter sets the shape of the mesh. If the value is -1 (default value) then <a href="https://en.wikipedia.org/wiki/Delaunay_triangulation">Delaunay triangulation</a> is used. If &gt;0 then the <a href="https://en.wikipedia.org/wiki/Alpha_shape">alpha-shape algorithm</a> is used. If 0, the <a href="https://en.wikipedia.org/wiki/Convex_hull">convex hull</a> is represented (resulting in a convex body).</p>


```

import plotly.graph_objects as go
import numpy as np

pts = np.loadtxt(np.DataSource().open('https://raw.githubusercontent.com/plotly/datasets/master/mesh_dataset.txt'))
x, y, z = pts.T

fig = go.Figure(data=[go.Mesh3d(x=x, y=y, z=z,
                   alphahull=5,
                   opacity=0.4,
                   color='cyan')])
fig.show()

```


<h3 id="Mesh-Tetrahedron">Mesh Tetrahedron<a class="anchor-link" href="#Mesh-Tetrahedron">¶</a></h3><p>In this example we use the <code>ì</code>, <code>j</code> and <code>k</code> parameters to specify manually the geometry of the triangles of the mesh.</p>


```

import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Mesh3d(
        x=[0, 1, 2, 0],
        y=[0, 0, 1, 2],
        z=[0, 2, 0, 1],
        colorbar_title='z',
        colorscale=[[0, 'gold'], 
                    [0.5, 'mediumturquoise'], 
                    [1, 'magenta']],
        # Intensity of each vertex, which will be interpolated and color-coded
        intensity=[0, 0.33, 0.66, 1],
        # i, j and k give the vertices of triangles
        # here we represent the 4 triangles of the tetrahedron surface
        i=[0, 0, 0, 1],
        j=[1, 2, 3, 2],
        k=[2, 3, 1, 3],
        name='y',
        showscale=True
    )
])

fig.show()

```


<h3 id="Mesh-Cube">Mesh Cube<a class="anchor-link" href="#Mesh-Cube">¶</a></h3>


```

import plotly.graph_objects as go
fig = go.Figure(data=[
    go.Mesh3d(
        # 8 vertices of a cube
        x=[0, 0, 1, 1, 0, 0, 1, 1],
        y=[0, 1, 1, 0, 0, 1, 1, 0],
        z=[0, 0, 0, 0, 1, 1, 1, 1],
        colorbar_title='z',
        colorscale=[[0, 'gold'],
                    [0.5, 'mediumturquoise'], 
                    [1, 'magenta']],
        # Intensity of each vertex, which will be interpolated and color-coded
        intensity = np.linspace(0, 1, 8, endpoint=True),
        # i, j and k give the vertices of triangles
        i = [7, 0, 0, 0, 4, 4, 6, 6, 4, 0, 3, 2],
        j = [3, 4, 1, 2, 5, 6, 5, 2, 0, 1, 6, 3],
        k = [0, 7, 2, 3, 6, 7, 1, 1, 5, 5, 7, 6],
        name='y',
        showscale=True
    )
])

fig.show()

```
